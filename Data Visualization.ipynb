Alternat Hypothesis H1 : Avg urban unemployment rate are > then rural unemployment rate
AS we can see Urban Unemploment avg rate is > then rural avg, So we can say that urban areas faces more unemployment then ruela areas, 
As result we fail to reject the alternat hypothesis H1
**Insight:** Urban unemployment rates are on average 3% higher than rural areas.
